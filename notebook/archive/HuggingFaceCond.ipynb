```python
tokenizer = CTRLTokenizer.from_pretrained('ctrl')
```

```python
model = CTRLLMHeadModel.from_pretrained('ctrl')
```

```python
model
```

```python
# inputs = tokenizer("Feminism", return_tensors="pt")
encoded_prompt = tokenizer.encode("Finance", add_special_tokens=False, return_tensors="pt")

output_sequences = model.generate(
    input_ids=encoded_prompt,
    max_length=80,
    temperature=0.5,
    top_k=0,
    top_p=0.9,
    repetition_penalty=1.2,
    do_sample=True,
    num_return_sequences=5
)

texts = [tokenizer.decode(sequence.tolist(), clean_up_tokenization_spaces=True) for sequence in output_sequences]
```
